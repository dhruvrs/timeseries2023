```python
import pandas as pd
import numpy as np

df = pd.read_csv('C:/Users/dhruv/Documents/School/Time Series/code/updateBox.csv')
cpi = pd.read_csv('C:/Users/dhruv/Documents/School/Time Series/code/cpi.csv')
df
```

```python
# put the cpi data into the box office dataframe
df['cpi'] = cpi['CORESTICKM159SFRBATL']

# converting all the nominal values to real using the percent change cpi data
df['Real Value for Overall'] = df.apply(lambda x: x['Overall Gross'] / (1 + x['cpi'] / 100), axis=1)
df['Real Value for Top 10'] = df.apply(lambda x: x['Top 10 Gross'] / (1 + x['cpi'] / 100), axis=1)
df.to_csv('finalData.csv', index=False)
df
```
