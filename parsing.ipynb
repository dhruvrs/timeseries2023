```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from fredapi import Fred
import warnings
warnings.filterwarnings('ignore')
```

```python
quarters = ['q1', 'q2', 'q3', 'q4']
urls = [f'https://www.boxofficemojo.com/quarter/{q}/?grossesOption=calendarGrosses' for q in quarters]
dfs = [pd.read_html(url, header=0)[0] for url in urls]
dfs[1]
```

```python
import re

for i in range(len(dfs)):
    dfs[i].sort_index(ascending=False, inplace=True)
    dfs[i]['Cumulative Gross'] = dfs[i]['Cumulative Gross'][::-1].reset_index(drop=True)
    dfs[i]['Year'] = dfs[i]['Year'][::-1].reset_index(drop=True)
    dfs[i]['Cumulative Gross'] = dfs[i]['Cumulative Gross'].apply(lambda x: re.sub('[^0-9]', '', str(x))) 
print(dfs[3]['Cumulative Gross'])
```

```python
# Create a date range starting from 2000 with quarterly frequency
dates = pd.date_range(start='1/1/2000', end='1/1/2023', freq='Q')

pretty = pd.DataFrame({
    'q1 gross': dfs[0]['Cumulative Gross'].reset_index(drop=True),
    'q2 gross': dfs[1]['Cumulative Gross'].reset_index(drop=True),
    'q3 gross': dfs[2]['Cumulative Gross'].reset_index(drop=True),
    'q4 gross': dfs[3]['Cumulative Gross'].reset_index(drop=True)
})

# pretty['q1 gross'] = pretty["q1 gross"].drop(pretty.index[:23], inplace=True)
# Create a DataFrame with the quarterly dates
# pretty = pd.DataFrame(dates, columns=['quarterly_date'])
pretty.to_csv('terror.csv')
# Print the DataFrame
```
